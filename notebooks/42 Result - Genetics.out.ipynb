## Phylogeny

-   A. puertoricana, A. galaica, A. eleganta are the same species - confirmed by COI, 18S, proteomics. Morphologies are also similar.
-   18S showed that *Acanthocope* is not a monophyly genus. COI disagrees.
-   Osborn ([2009](#ref-osborn2009)) has already mentioned a potential revision/unresolved relationship between Acanthocopinae and Storyhyngurinae, but mentioned that the same topology was not reproducible in raupach2009, which became a polytomy in their 18S tree.
-   malyutina2003 revised the Storthyngurinae in morphology and the revision is confirmed by Raupach et al. ([2009](#ref-raupach2009)) (double check this).

Storthyngurinae (*Storthyngura, Storthyngurella, Sursumura, Vanhoeffenella, Rectisura*) might be a group nested inside *Acanthocope*. I got a specimen of *Sursumura sp.* (10.3.25, S0803) and it looks very, very, very similar to *Acanthocope* except P1 dorsomedial spine and pleotelson. I am still improving 18S and COI analysis to confirm this. My current version of 18S phylogeny has relative high confidence on this.
